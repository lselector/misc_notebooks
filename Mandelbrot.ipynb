### Mandelbrot Set

from:
<br> - https://github.com/danyaal/mandelbrot/blob/master/mandelbrot.py
```
Consider function f(z)=z*z + c, where c is a complex number.
We create a sequence:
  z0 = 0
  z1 = f(z0)
  z2 = f(z1)
  etc.
The Mandelbrot set is the set of complex numbers "c"
for which the above process does not diverge 
(does not go to infinity).
```

````python
import os, sys, time, random, numpy as np
# from IPython.display import clear_output
from matplotlib import pyplot as plt
# %config InlineBackend.figure_format = 'retina'
````

````python
def countIterationsUntilDivergent(c, threshold):
    """
    # counts the number of iterations until the function diverges or
    # returns the iteration threshold that we check until
    """
    z = complex(0, 0)
    for iteration in range(threshold):
        z = (z*z) + c

        if abs(z) > 4:
            break
            pass
        pass
    return iteration
````

````python
def mandelbrot(threshold, density, xr=None, yr=None):
    """
    # This function creates atlas, then plots the mandelbrot set
    # Parameters:
    #    threshold - iteration limit before declaring function as convergent 
    #    density (of the atlas)
    #    xr, yr - size of the atlas rectangle
    """
    # location and size of the atlas rectangle
    realAxis      = np.linspace(xr[0], xr[1], density)
    imaginaryAxis = np.linspace(yr[0], yr[1], density)

    realAxisLen = len(realAxis)
    imaginaryAxisLen = len(imaginaryAxis)

    # 2-D array to represent mandelbrot atlas
    atlas = np.empty((realAxisLen, imaginaryAxisLen))

    # color each point in the atlas depending on the iteration count
    for ix in range(realAxisLen):
        for iy in range(imaginaryAxisLen):
            cx = realAxis[ix]
            cy = imaginaryAxis[iy]
            c = complex(cx, cy)

            atlas[ix, iy] = countIterationsUntilDivergent(c, threshold)
            pass
        pass

    # plot and display mandelbrot set
    plt.imshow(atlas.T, interpolation="nearest")
    plt.show()
````

````python
# calculate and show - takes less than 10 seconds
mandelbrot(threshold=120, density=1000, xr=[-2.25,  0.75 ], yr=[-1.5 ,  1.5  ] )
````
